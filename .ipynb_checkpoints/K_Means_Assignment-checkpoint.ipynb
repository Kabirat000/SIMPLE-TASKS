**Case study** of training and tuning a k-means clustering model using a real-world California housing dataset from [Kaggle](https://www.kaggle.com/datasets/camnugent/california-housing-prices?resource=download).

**Goal**: We want to analyze how house prices vary across different locations in California by clustering houses based on their **latitude** and **longitude** coordinates, along with the **median house value**.

We'll use location data (latitude and longitude) and median house value to cluster the houses, allowing us to identify patterns and trends in house prices across different areas of California.

```python
import pandas as pd

df = pd.read_csv('housing.csv')

home_data = df[['longitude', 'latitude', 'median_house_value']]
home_data.head()
```

## Tasks:
- Split the dataset, where features (X) are `home_data[['latitude', 'longitude']]` and target (y) is `home_data['median_house_value']`.
- Normalize the features (X) data using the `preprocessing.normalize()` method from sklearn.
- Apply KMeans clustering to the normalized dataset.
- Train the model and obtain cluster labels for each data point.
- Visualize the data we just fit using scatter plot. Hint: `hue = kmeans.labels_`.
- Write your observation.
- Plot the distribution of median house prices in the number of groups using a boxplot.
Hint: `sns.boxplot(x = kmeans.labels_, y = home_data['median_house_value'])`
- Can you choose the best number of clusters using the elbow method and the silhouette score?

```python
# Split the dataset
X = home_data[['latitude', 'longitude']]
y = home_data['median_house_value']
```

```python
# Normalize the features (X)
from sklearn import preprocessing

X_norm = preprocessing.normalize(X)
```

- When working with distance-based algorithms such as KMeans Clustering, it is crucial to normalize the data to ensure that all features contribute equally to the distance computations.
- If the data is not normalized, features with larger scales will disproportionately influence the distance metric, leading to suboptimal clustering results.

**Fitting and Evaluating the Model**
- For the first iteration, we'll arbitrarily choose 3 clusters (k=3).

```python
# Apply KMeans clustering to the normalized dataset
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(X_norm)
```

After fitting the data, we can access the labels from the `labels_` attribute. Below, we visualize the fitted data.

```python
# Visualize the data we just fit using scatter plot
sns.scatterplot(data = X, x = 'longitude', y = 'latitude', hue = kmeans.labels_)
```

**Observation:**

The data is now clearly split into three groups say, Northern California, Central California, and Southern California.

```python
# Plot the distribution of median house prices in the number of groups using a boxplot

sns.boxplot(x = kmeans.labels_, y = y)
```

The Central and Southern clusters (clusters 0 and 1) have similar, higher median house values compared to the Northern cluster (cluster 2).

More info about [boxplot](https://www.wellbeingatschool.org.nz/information-sheet/understanding-and-interpreting-box-plots).
____

We can evaluate the performance of the clustering algorithm using the Silhouette score from `sklearn.metrics`, where a higher score indicates a better fit.

```python
from sklearn.metrics import silhouette_score

silhouette_score(X_norm, kmeans.labels_, metric='euclidean')
```

So far, we haven't assessed the effectiveness of various cluster numbers, hence, we're uncertain about how well the *k = 3* model fits.

Next, we'll investigate different cluster options and compare their performance to determine the optimal hyperparameter values for our model.

```python
k_values = range(2, 8)
fits = []
scores = []


for k in k_values:
    # train the model for current value of k on X_norm
    model = KMeans(n_clusters = k, random_state = 0, n_init='auto').fit(X_norm)

    # append the model to fits
    fits.append(model)

    # Append the silhouette score to scores
    scores.append(silhouette_score(X_norm, model.labels_, metric='euclidean'))
```

```python
scores
```

```python
fits
```

Let's visually look at a few different values of k.

```python
# k = 2
sns.scatterplot(data = X, x = 'longitude', y = 'latitude', hue = fits[0].labels_)
```

The model somewhat divides the state into two halves, but it may not capture enough subtleties in the California housing market.

```python
# k = 4
sns.scatterplot(data = X, x = 'longitude', y = 'latitude', hue = fits[2].labels_)
```

In this case, this plot groups California into more logical clusters based on their location in the state, either North or South. This model likely captures more subtleties in the housing market as we move across California.

```python
#k = 7
sns.scatterplot(data = X, x = 'longitude', y = 'latitude', hue = fits[5].labels_)
```

The graph above seems to have an excessive number of clusters. We've traded off easy cluster interpretation for a "more accurate" geographical clustering outcome.

```python
sns.lineplot(x = k_values, y = scores)
```

We usually select the point where the performance improvements start to level off or degrade. In this case, k = 5 appears to be the optimal choice without risking overfitting.

```python
sns.scatterplot(data = X, x = 'longitude', y = 'latitude', hue = fits[3].labels_)
```

```python
sns.boxplot(x = fits[3].labels_, y = y)
```

We can observe that the clusters effectively divide California into distinct regions, and these regions correspond reasonably well to different price ranges.
